# Load results and explore

```r
result = read.csv('cellphoneDB/out_Rversion/relevant_interactions.txt', stringsAsFactors = F, sep = '\t')
```

```r
cell1 = 'Germ.PGC'
cell2 = 'Somatic.CoelEpi_LHX9'
combibations = c(paste(cell1, cell2, sep='...'), paste(cell2, cell1, sep='...'))

idx_cl = which(colnames(result) %in%  combibations) 
idx_row = result[combibations[1]] == 1 | result[combibations[2]] == 1

x = subset(result, idx_row )[, c(3:6, idx_cl)]
# x
genes = c( 'SIRPA')  
subset(x, gene_b %in% genes | gene_a %in% genes)
```

```r
EMC_germ = c('ITGA6', 'ITGB1', 
             'CADM3',  'LAMC1',
               'JAM2', 'CD47',
              'COL4A1', 'COL4A2',  'COL9A3',
             'AGRN', 'COL6A2',  
             "NECTIN3", 'COL1A1', "ITGAV", 'COL3A1', 'COL1A2',
             'ITGA4',  'FN1',  'CDH1', "ITGA11")       


EMC_soma = c('ITGB4', 'COL5A2', 'COL8A1', 
             'COL16A1','COL9A3',   'COL27A1', 
            'ITGA5', 'JAM2', 'JAM3',  'SIRPA', 
            'COL11A1', 'CADM1',
             'COL4A5', 'COL14A1',  'COL6A2', 'COL6A3', 'COL3A1', 'COL4A1', 'COL4A2',"COL4A3", "COL4A4", 'COL6A1', 'COL21A1', 
             'ITGA6', 
            'ITGB1',   
            'LAMC1',   "ITGAV",  'ITGA2'
#             'ITGA8'
           )
genes = c(EMC_germ, EMC_soma)
X = subset(result, gene_b %in% genes | gene_a %in% genes)
X = X[ ,11:ncol(X)]
rownames(X) = subset(result, gene_b %in% genes | gene_a %in% genes)$interacting_pair
```

```r
ECM_df = melt(as.matrix(X))
ECM_df = subset(ECM_df, value == 1)
ECM_df = ECM_df[ intersect(grep('Soma', ECM_df$Var2), grep('Germ', ECM_df$Var2)) , ]
```

```r
cell1 = 'Germ.pre_oocyte'
cell2 = 'Somatic.granulosa'
combibations = c(paste(cell1, cell2, sep='...'), paste(cell2, cell1, sep='...'))

subset(ECM_df, Var2 %in% combibations)
```
